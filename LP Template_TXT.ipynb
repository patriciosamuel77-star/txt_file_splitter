```python
import pandas as pd
from pathlib import Path

excel_path = Path("Template TXT prueba.xlsx")

df = pd.read_excel(
    excel_path,
    sheet_name=0,
    keep_default_na=False,
    dtype=str
)

# limpiar encabezados solamente
df.columns = df.columns.str.strip()

# validar columnas
required = {"COMPANY CODE", "IMPORT"}
missing = required - set(df.columns)
if missing:
    raise ValueError(f"Faltan columnas en Excel: {missing}")

df = df[["COMPANY CODE", "IMPORT"]]

# limpiar solo COMPANY CODE
df["COMPANY CODE"] = df["COMPANY CODE"].str.strip()

# eliminar solo filas donde ambas columnas estén vacías
df = df[~((df["COMPANY CODE"] == "") & (df["IMPORT"] == ""))]

output_dir = excel_path.parent

for company_code, df_cc in df.groupby("COMPANY CODE", sort=False):

    file_name = f"02.26 - Importacion {company_code}.txt"
    output_file = output_dir / file_name

    with open(output_file, "w", encoding="utf-8") as f:
        for line in df_cc["IMPORT"]:
            f.write(line + "\n")

    print(f"Generado: {output_file}")
```

```python
import zipfile
from pathlib import Path

zip_name = "TXT_por_company_code.zip"
periodo = "02.26"

with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zipf:
    for company_code in df["COMPANY CODE"].unique():
        file_name = f"{periodo} - Importacion {company_code}.txt"
        txt_file = Path(file_name)

        if txt_file.exists():
            zipf.write(txt_file)

print("ZIP generado")
```
